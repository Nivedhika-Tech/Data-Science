```python
import pandas as pd
import scipy.stats as stats

df2['class'] = df2['class'].astype('category')

Economy = df2[df2['class'] == 'Economy']['price']
Business = df2[df2['class'] == 'Business']['price']

ttest_result = stats.ttest_ind(Economy, Business)

print(f"T-statistic: {ttest_result.statistic}")
print(f"P-value: {ttest_result.pvalue}")

if ttest_result.pvalue < 0.05:
    print("There is a significant difference in price between Economy and Business class")
else:
    print("There is no significant difference in price between Economy and Business class")
```

```python
import pandas as pd
import scipy.stats as stats

df2['airline'] = df2['airline'].astype('category')

SpiceJet = df2[df2['airline'] == 'SpiceJet']['price']
AirAsia = df2[df2['airline'] == 'AirAsia']['price']
Vistara = df2[df2['airline'] == 'Vistara']['price']
GO_FIRST = df2[df2['airline'] == 'GO_FIRST']['price']
Indigo = df2[df2['airline'] == 'Indigo']['price']
Air_India = df2[df2['airline'] == 'Air_India']['price']

anova_result = stats.f_oneway(SpiceJet, AirAsia, Vistara, GO_FIRST, Indigo, Air_India)

print(f"F-statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")

if anova_result.pvalue < 0.05:
    print("There is a  difference in price between airline.")
else:
    print("There is no significant difference in price between airline")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

table = pd.crosstab(df2['airline'], df2['class'])


chi2_stat, p_val, dof, expected = chi2_contingency(table)
print( chi2_stat)
print(p_val)

if p_val < 0.05:
    print("There is difference between airline and class")  # there is relationship
else:
    print("There is no difference between airline and class") # there is no relationship
```
